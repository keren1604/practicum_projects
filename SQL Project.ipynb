**<font color='BlueViolet'>Find the number of user reviews and the average rating for each book.**



In this section I will create 3 tables:

-The first is arranged in descending order by number_of_reviews+ average_rating

-The second is arranged in descending order only by number_of_reviews

-The third is arranged in descending order only according to average_rating

This is so that I can examine in more depth what the leading books are

```python
query= '''
SELECT
    books.book_id,
    books.title,
    COUNT(DISTINCT(review_id)) AS number_of_reviews,
    AVG(rating) AS average_rating
        
FROM books
INNER JOIN reviews USING (book_id)
INNER JOIN ratings USING (book_id)

GROUP BY book_id
ORDER BY number_of_reviews DESC, average_rating DESC
LIMIT 10; '''

pd.io.sql.read_sql(query, con = engine)
```

**<font color='BlueViolet'>Identify the publisher that has released the greatest number of books with more than 50 pages (this will help you exclude brochures and similar publications from your analysis).**

```python
query= '''
SELECT books.publisher_id,
    publisher,
    COUNT(title) AS number_of_books
    
FROM books
INNER JOIN publishers USING (publisher_id)

WHERE books.num_pages >=50 
GROUP BY publisher, publisher_id
ORDER BY number_of_books DESC
LIMIT 1; '''

pd.io.sql.read_sql(query, con = engine)
```

**The publisher that has released the greatest number of books with more than 50 pages is Penguin Books, id=212, numbers of books= 42**

**<font color='BlueViolet'>Identify the author with the highest average book rating (look only at books with at least 50 ratings).**


```python
query= '''
SELECT authors.author,
    AVG(ratings.rating) AS average_rating
    
FROM
    books
    JOIN authors ON books.author_id= authors.author_id
    JOIN ratings ON books.book_id= ratings.book_id  
    
WHERE
    books.book_id IN (
        SELECT
            book_id
        FROM
            ratings
        GROUP BY
            book_id
        HAVING 
            COUNT(1) >=50
    )
GROUP BY
    authors.author
ORDER BY
    average_rating DESC
LIMIT 1; '''
pd.io.sql.read_sql(query, con = engine)
```

**The author with the highest average book rating (from books with 50 or more ratings) is J.K. Rowling/Mary GrandPré, having 4.287097 average rating**

**<font color='BlueViolet'>Find the average number of text reviews among users who rated more than 50 books.**

```python
query= '''
SELECT AVG(COUNT(DISTINCT(reviews.review_id))) OVER () AS average_number_of_reviews

FROM reviews
INNER JOIN ratings ON reviews.username=ratings.username
GROUP BY ratings.username
HAVING COUNT(DISTINCT(ratings.rating_id)) > 50
LIMIT 1; '''

pd.io.sql.read_sql(query, con = engine)
```

**The average number of text reviews among users who rated more than 50 books is 24.333 rating per user.**
